## Modeling

Let's start by modeling our monkeys.

Each `Monkey` has a number (its name), items it has stolen, operation that defines the new worry level, a condition that determines who to throw the item to and a count of how many it has inspected.

A `Condition` stores its divider and target monkey's number and evaluates a value given to it to decide where to throw it.

````python
class Monkey:
    
    def __init__(self, number, items, operation, condition):
        self.number = number
        self.items = items
        self.operation = operation
        self.condition = condition
        self.inspected = 0
        
    def __repr__(self):
        return f'<Monkey {self.number}: {self.items} | Ins: {self.inspected}>'
    
class Condition:
    
    def __init__(self, divider, if_true, if_false):
        self.divider = divider
        self.if_true = if_true
        self.if_false = if_false
        
    def evaluate(self, value):
        if value % self.divider == 0:
            return self.if_true
        else:
            return self.if_false
````

## Read input

The input is a multi-part input that gets split by empty lines ('\n\n') and then each part gets parsed into a `Monkey` and `Condition` objects.

As a result, we get a list of monkeys as our starting point.

````python
from utils import read_multisection_input
import re

monkeys_input = read_multisection_input(11, transformers=str)
example_input = read_multisection_input(11, transformers=str, example=True)

def parse_monkey(data):
    lines = data.split('\n')
    
    number = int(re.search(r'\d+', lines[0]).group())
    items = [int(item) for item in re.findall('\d+', lines[1])]
    operation = lines[2].split('= ')[-1]
    
    test_division = int(re.search(r'\d+', lines[3]).group())
    if_true = int(re.search(r'\d+', lines[4]).group())
    if_false = int(re.search(r'\d+', lines[5]).group())
    condition = Condition(test_division, if_true, if_false)
    
    return Monkey(number, items, operation, condition)
    
monkeys = [parse_monkey(d) for d in monkeys_input]
````

## Monkey simulations

This solution is generic and solves both part 1 and 2.

Our `simulate_round` gets a list of `monkeys`, a `divider` (needed for part 2, I'll explain it later) and `chill` flag that defines if our worry levels drop when monkey is done (for part 2).

For each monkey, we go through all the items, pick up the first one, evaluate its operation ([eval is dangerous](https://softwareengineering.stackexchange.com/a/311510/87030) but here we have a semitrusted input so I'm taking a shortcut).

If we're in chill mode (solving part 1), we divide the worry by 3 and round down (Python's `//` operator is "divide and convert to integer" compared to `/` which returns a float).

To keep Part 2's wildly increasing numbers in check, we take a modulo of the new value. We use [modular arithmetic](https://en.wikipedia.org/wiki/Modular_arithmetic) here and the divider is the [least common multiple](https://en.wikipedia.org/wiki/Least_common_multiple) of all the dividers in each monkey's conditions. I would have never figured this out myself and it was explained to me in one of our community groups where we solve Advent of Code puzzles and share solutions and help each other out. Thanks Toni for the help here!

Once we've done all the _math_, we add it to the target monkey's list and move on to the next item.

Because we're using `.pop(0)` here, I needed to make a copy of the list for the for loop as we're modifying the list on the fly. 

To simulate the entire thing, we first get the divider, then loop for N rounds (20, 10k respectively) and return the final state of the monkeys.

To calculate the monkey business score, we sort the monkeys by their `inspected` count and multiple two largest ones together.

````python
from math import lcm

def simulate_round(monkeys, divider, chill):
    for monkey in monkeys:       
        for old in monkey.items.copy():
            monkey.inspected += 1
            monkey.items.pop(0)
            new = eval(monkey.operation)  # Kids, don't do eval in real life. See https://softwareengineering.stackexchange.com/a/311510/87030
            if chill:
                new = new // 3
            new = new % divider
            target = monkey.condition.evaluate(new)
            monkeys[target].items.append(new)

def simulate_monkeys(monkeys, chill, rounds=20):
    divider = lcm(*[monkey.condition.divider for monkey in monkeys])
    for i in range(rounds):
        simulate_round(monkeys, divider, chill)
    return monkeys

def calculate_monkey_business(monkeys):
    monkeys = sorted(monkeys, key=lambda m: m.inspected)
    return monkeys[-1].inspected * monkeys[-2].inspected
````

````python
from copy import deepcopy

part1 = simulate_monkeys(deepcopy(monkeys), True)

p1 = calculate_monkey_business(part1)

print(f'Part 1: {p1}')
assert p1 == 66124
````
